```python
'''
    Get last 140_000 bars, end date presented here
'''

months = {
    'January':'F', 'February':'G', 'March':'H',
    'April':'J', 'May':'K', 'June':'M',
    'July':'N', 'August':'Q', 'September':'U',
    'October':'V', 'November':'X', 'December':'Z',
}
m = list(months.keys())


end_year = 2022
years = 50

contract_name = 'CL'
contract_interval = 1

for year in range(years):
    contract_year = str(end_year - year)[2:]
    for n, (month_name, month_letter) in enumerate(reversed(months.items())):
        if contract_interval == 3:
            if month_name in ('December', 'September', 'June', 'March'):
                bc_name = f"{contract_name}{month_letter}{contract_year}"
            else:
                continue
        else:
            bc_name = f"{contract_name}{month_letter}{contract_year}"
        get_month = m[-n]
        get_year = f"20{contract_year}" if get_month != 'January' \
                   else f"20{int(contract_year)+1}"

        print(f"{month_name[:3]}-{contract_year} {bc_name}, end: {m.index(get_month)+1}/1/{get_year}")
    
```
